```python
import turtle
import math
class circle:
    def __init__(self, r, x, y):
        self.r = r
        self.x = x
        self.y = y
    def draw(self):
        t =turtle.Turtle()
        t.hideturtle()
        t.penup()
        t.goto(self.x, self.y)
        t.pendown()
        t.circle(self.r)
        turtle.done()
    def area(self):
        return math.pi * self.r ** 2
    def perimeter(self):
        return 2 * math.pi * self.r
c = circle(100, -200, 0)
c.draw()
print("S = ", c.area())
print("C = ", c.perimeter())
```
